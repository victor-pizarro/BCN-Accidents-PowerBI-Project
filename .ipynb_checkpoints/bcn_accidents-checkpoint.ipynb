```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
#!pip install utm : librearia para pasar de coordenadas utm a longitud&latitud
import utm
```

```python
pd.set_option('max_columns', None)
df=pd.read_csv('dataset/accidents_opendata.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# Eliminar columnas no útiles
df.drop(['case_id',
         'district_id',
         'neighborhood_id',
         'street_code',
         'street_name',
         'postal_code',
         'weekday',
         'month',
         'day',
         'longitude',
         'latitude',
         'type_day'],
         axis=1,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
# Revisamos la cantidad de duplicados.
df.duplicated().sum()
```

```python
#Eliminamos duplicados
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
# Revisamos la cantida de nulos
df.isnull().sum()
```

```python
#Eliminamos valores nulos
df=df.dropna()
df.isnull().sum()
```

```python
#Cambiamos los data type de float a int
df['year']=df['year'].astype(int)
df['n_deaths']=df['n_deaths'].astype(int)
df['n_wounded_mild']=df['n_wounded_mild'].astype(int)
df['n_wounded_severe']=df['n_wounded_severe'].astype(int)
df['n_victims']=df['n_victims'].astype(int)
df['n_vehicles']=df['n_vehicles'].astype(int)
df['hour']=df['hour'].astype(int)
df['utm_coordinate_y']=df['utm_coordinate_y'].apply(lambda x: float(x.replace(',','.')))
df['utm_coordinate_x']=df['utm_coordinate_x'].apply(lambda x: float(x.replace(',','.')))
df.info()
```

```python
# Reseteamos el indice, al eliminar registros/filas también se eliminan los indices.
df=df.reset_index()
```

```python
# Como obtener latitud y longitud a partir de coordenadas UTM
# Cataluña: 31 , T (Referencia según medidas UTM para posicionar Cataluña)
# utm.to_latlon(utm_coordinate_x, utm_coordinate_y, 31, 'T')

# Coordenadas
print(utm.to_latlon(431779.16, 4585420.58, 31, 'T'))

# Latitud
print(utm.to_latlon(431779.16, 4585420.58, 31, 'T')[0])

# Longitud
print(utm.to_latlon(431779.16, 4585420.58, 31, 'T')[1])
```

```python
df.shape
```

```python
# Creamos una nueva columna con las latitudes
latitude=[]
for i in range(0,df.shape[0]):
    try:
        latitude.append(utm.to_latlon(df.iloc[i]['utm_coordinate_x'],df.iloc[i]['utm_coordinate_y'],31,'T')[0])
    except:
        latitude.append(np.nan)
df['latitude']=latitude
```

```python
# Creamos una nueva columna con las longitudes
longitude=[]
for i in range(0,df.shape[0]):
    try:
        longitude.append(utm.to_latlon(df.iloc[i]['utm_coordinate_x'],df.iloc[i]['utm_coordinate_y'],31,'T')[1])
    except:
        longitude.append(np.nan)
df['longitude']=longitude
```
